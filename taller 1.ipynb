```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df
```

```python
promedio_edad = df['Age'].mean()
promedio_edad
```

```python
valores_nulos_por_columna = df.isnull().sum()
valores_nulos_por_columna
```

```python
genero_mas_repetido =df['Gender'].mode()
genero_mas_repetido
```

```python
hombres = (df['Gender']==0).sum()
hombres
```

```python
#solucion taller
#cantidad de mujeres
mujeres = (df['Gender']==1).sum()
mujeres
```

```python
#cantidad de estudiantes que practican deportes
estudiantes_practican_deporte = df['Sports'].sum()
estudiantes_practican_deporte
```

```python
#cantidad de hombres y mujeres que practican deporte
hombres_deportistas = df[(df['Gender'] ==0 ) & (df['Sports'] == 1)]['Sports'].sum()
mujeres_deportistas = df[(df['Gender']==1 ) & (df['Sports']== 1)]['Sports'] . sum()

print ("la catindad de hombres que practican deporte son: ", hombres_deportistas)
print ("la cantidad de mujeres que practican deporte son: ", mujeres_deportistas)
```

```python
#cantidad de personas que van tutorias
estudiantes_asisten_tutorias = df['Tutoring'].sum()
print ("la cantida de estudiantes que asisten a tutorias son: ", estudiantes_asisten_tutorias)
```

```python
# cantidad de hombres y mujeres que van a tutorias.
tutorias_genero = df.groupby('Gender')['Tutoring']. sum()
tutorias_genero
```

```python
# Tiempo de estudio max y minimo por semana
tiempo_max = df['StudyTimeWeekly'].max()
tiempo_min = df['StudyTimeWeekly'].min()

print("el tiempo maximo de estudio por semana es: ", tiempo_max)
print("el tiempo minimo de estudio por semana es: ", tiempo_min)
```
